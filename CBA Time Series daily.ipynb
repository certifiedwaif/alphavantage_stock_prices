```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import stock_prices
```

```python
data = stock_prices.get_stock_prices()
```

```python
time_series_daily = data['Time Series (Daily)']
```

```python
pd.DataFrame?
```

```python
records = []
for date, d in time_series_daily.items():
    d.update({'date' : date})
    records.append(d)
df = pd.DataFrame.from_records(records)
df.keys()
```

```python
df = df.rename(index=str, columns={
    '1. open' : 'open',
    '2. high' : 'high',
    '3. low' : 'low',
    '4. close' : 'close',
    '5. adjusted close' : 'adjusted_close',
    '6. volume' : 'volume',
    '7. dividend amount' : 'divident_amount',
    '8. split coefficient' : 'split_coefficient'
})
df.head()
```

```python
plt.plot(df.open)
plt.plot(df.high)
plt.plot(df.low)
plt.plot(df.close)
plt.plot(df.adjusted_close)
plt.legend()
plt.title('CBA Time Series daily')
```

```python
plt.plot(df.open, df.close, "o")
plt.xlabel('open')
plt.ylabel('close')
plt.title('CBA open versus close prices')
```
