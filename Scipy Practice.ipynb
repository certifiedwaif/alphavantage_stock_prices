```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import scipy.stats as stats
```

```python
mu = [0., 0.]
sigma = [[1., 0.3],
         [0.3, 1.]]
print(mu)
print(sigma)
mvn = stats.multivariate_normal(mean=mu, cov=sigma)
x = np.linspace(-2, 2)
y = np.linspace(-2, 2)
X, Y = np.meshgrid(x, y)
pos = np.dstack((X, Y))
fig = plt.figure()
ax = fig.add_subplot(111)
ax.contourf(x, y, mvn.pdf(pos))
```
